## II. **EDA (Exploratory Data Analysis)**

To identify potential predictors of financial shocks, we begin with a correlation analysis between stock market indices and key macroeconomic indicators.

### **Key Findings:**
✔ **GDP and Inflation have a strong positive correlation with stock market performance, making them key leading indicators.**  
✔ **Unemployment and Interest Rates exhibit a negative correlation with stock indices, with unemployment serving as a potential predictor of downturns.**  
✔ **Nasdaq appears more sensitive to fluctuations in unemployment compared to the S&P 500.**  
✔ **VIX effectively captures periods of market panic but lacks strong predictive power for long-term trends.**  

### **1️⃣ Compute Correlation Matrix**

The correlation matrix shows how strongly different variables are related:
- **S&P 500 (GSPC) and Nasdaq (IXIC) have a strong positive correlation** with **Inflation and GDP** (~0.9).  
  - This suggests that rising GDP and inflation often coincide with rising stock prices.
- **Unemployment has a moderate negative correlation** with stock indices (~ -0.4).  
  - When unemployment rises, stock prices tend to fall.
- **VIX (Volatility Index) has weak negative correlations** with stock indices (~ -0.1 to -0.2).  
  - While VIX spikes during market crashes, it does not consistently predict downturns.

```python
df_financial = df.copy()
```

```python
# Reset the index and store it in a column called 'Date'
df_financial = df_financial.reset_index().rename(columns={'index': 'Date'})

# Convert 'Date' to datetime format
df_financial['Date'] = pd.to_datetime(df_financial['Date'], errors='coerce')

# Confirm the changes
df_financial.head()
```

```python
# Select key financial and macroeconomic indicators
correlation_cols = ['Adj Close_^GSPC', 'Adj Close_^IXIC', 'Adj Close_^VIX', 
                    'Bond Yields', 'Inflation', 'Unemployment', 'Interest Rate', 'Consumer Sentiment', 'GDP']

# Filter dataset to only include available columns
df_corr = df_financial[correlation_cols].dropna()

# Compute correlation matrix
corr_matrix = df_corr.corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Stock Indices & Macroeconomic Indicators")
plt.show()
```

### **2️⃣ Identify Leading Indicators (Lagged Correlations)**

By introducing time lags, we can see if macroeconomic indicators **predict future market movements**:
- **GDP and Inflation remain strongly correlated with stocks (~0.92+), suggesting they are leading indicators**.
- **Unemployment shows a stronger negative correlation when lagged (~ -0.41)**, indicating **higher unemployment precedes stock market declines**.
- **Bond Yields & Interest Rates have stronger lagged correlations (~0.75)**, implying their effects on markets take time to materialize.

```python
# Shift macroeconomic variables to see how past values affect current stock indices
df_corr_shifted = df_financial.copy()
df_corr_shifted[['Bond Yields', 'Inflation', 'Unemployment', 'Interest Rate', 'GDP']] = df_corr_shifted[
    ['Bond Yields', 'Inflation', 'Unemployment', 'Interest Rate', 'GDP']
].shift(30)  # 30-day lag

# Compute new correlation matrix with lagged variables
lagged_corr_matrix = df_corr_shifted[correlation_cols].dropna().corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(lagged_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Lagged Correlation Matrix: Leading Indicators")
plt.show()
```

### **3️⃣ Scatter Plots for Key Relationships**

To further explore these relationships, we plot key economic variables against stock indices:
- **Interest Rates vs. S&P 500**
  - **Nonlinear relationship**:  
    - At **lower interest rates (0-2%)**, S&P 500 remains high.  
    - At **higher rates (>3%)**, stock prices tend to be lower or volatile.  
  - This suggests that **interest rate hikes may suppress stock market growth**.
- **Unemployment vs. Nasdaq**
  - **Clear negative correlation**: Nasdaq performs well when unemployment is low (~4-6%) but declines sharply when unemployment exceeds 8-10%.  
  - **Recession periods likely align with high unemployment**, impacting market performance.

```python
# Scatter plot: Interest Rate vs. S&P 500
plt.figure(figsize=(6,4))
sns.scatterplot(x=df_corr['Interest Rate'], y=df_corr['Adj Close_^GSPC'])
plt.title("Interest Rate vs S&P 500")
plt.xlabel("Interest Rate")
plt.ylabel("S&P 500 Closing Price")
plt.grid(True)
plt.show()

# Scatter plot: Unemployment vs. Nasdaq
plt.figure(figsize=(6,4))
sns.scatterplot(x=df_corr['Unemployment'], y=df_corr['Adj Close_^IXIC'])
plt.title("Unemployment Rate vs Nasdaq")
plt.xlabel("Unemployment Rate")
plt.ylabel("Nasdaq Closing Price")
plt.grid(True)
plt.show()
```

## 📊 Rolling Volatility Analysis

To better understand market **instability**, we analyze **rolling volatility** using a **30-day moving standard deviation** of stock indices.

### **Key Insights**
✔ **High-volatility periods align with major financial events:**
   - **2008 Global Financial Crisis** shows a **volatility spike**.
   - **COVID-19 market crash (2020)** caused extreme swings.
   - **Recent years (2022–2024)** show increased volatility, likely due to **inflation, interest rate hikes, or geopolitical events**.

✔ **Nasdaq (^IXIC) has the highest volatility**, meaning it's more sensitive to market shifts.

✔ **VIX (Volatility Index) remains low compared to stock indices**, reinforcing its role as a **short-term panic indicator rather than a long-term predictor**.

### **Next Steps**
📌 Do we want to:
- **Overlay recession periods on this volatility chart?**
- **Detect financial shock anomalies based on extreme volatility?**
- **Compare volatility to macroeconomic indicators (e.g., inflation, interest rates)?**

```python
# ✅ What This Does
# 	•	Computes rolling volatility (30-day standard deviation) for S&P 500, Nasdaq, and VIX.
# 	•	Plots market volatility over time to detect financial shocks.
# 	•	Highlights volatility spikes during crisis periods.

import matplotlib.pyplot as plt

# Define a rolling window size (e.g., 30 days for monthly volatility)
rolling_window = 30

# Select stock indices for volatility analysis
volatility_cols = ['Adj Close_^GSPC', 'Adj Close_^IXIC', 'Adj Close_^VIX']

# Compute rolling standard deviation (volatility)
df_volatility = df_financial[['Date'] + volatility_cols].copy()
for col in volatility_cols:
    if col in df_volatility.columns:
        df_volatility[f'Rolling_Volatility_{col}'] = df_volatility[col].rolling(window=rolling_window).std()

# Plot rolling volatility for stock indices
plt.figure(figsize=(14, 6))
for col in volatility_cols:
    if f'Rolling_Volatility_{col}' in df_volatility.columns:
        plt.plot(df_volatility['Date'], df_volatility[f'Rolling_Volatility_{col}'], label=f'Volatility {col}')

plt.title('Rolling Volatility of Stock Indices (30-Day Window)')
plt.xlabel('Date')
plt.ylabel('Rolling Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Select macroeconomic variables and volatility metrics
macro_cols = ['Bond Yields', 'Inflation', 'Unemployment', 'Interest Rate', 'GDP']
vol_cols = [f'Rolling_Volatility_{col}' for col in volatility_cols]
df_corr = df_volatility.merge(df_financial[['Date'] + macro_cols], on='Date', how='left')

# Compute correlation matrix
corr_matrix = df_corr[vol_cols + macro_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Volatility & Macroeconomic Indicators")
plt.show()
```

```python
# Compute Z-score for rolling volatility
for col in volatility_cols:
    df_volatility[f'Volatility_Zscore_{col}'] = (
        df_volatility[f'Rolling_Volatility_{col}'] - df_volatility[f'Rolling_Volatility_{col}'].mean()
    ) / df_volatility[f'Rolling_Volatility_{col}'].std()

# Define anomaly threshold (e.g., Z-score > 2)
threshold = 2
anomalies = df_volatility[df_volatility.filter(like='Volatility_Zscore').abs() > threshold]

# Plot volatility with anomalies
plt.figure(figsize=(14, 6))
for col in volatility_cols:
    plt.plot(df_volatility['Date'], df_volatility[f'Rolling_Volatility_{col}'], label=f'Volatility {col}')

# Highlight anomalies
plt.scatter(anomalies['Date'], anomalies['Rolling_Volatility_Adj Close_^GSPC'], color='red', label="Anomalies")

plt.title('Anomalous Volatility Events')
plt.xlabel('Date')
plt.ylabel('Rolling Standard Deviation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Select a volatility column (e.g., S&P 500)
vol_series = df_volatility.set_index('Date')['Rolling_Volatility_Adj Close_^GSPC'].dropna()

# Fit ARIMA model (auto-select parameters)
model = ARIMA(vol_series, order=(5,1,0))
model_fit = model.fit()

# Forecast next 30 days
forecast = model_fit.forecast(steps=30)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(vol_series, label="Historical Volatility")
plt.plot(forecast.index, forecast, label="Forecasted Volatility", color='red')
plt.title("Volatility Forecasting with ARIMA")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.legend()
plt.show()
```
