```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Load one or more synthetic datasets
df = pd.read_csv("../synthetic_datasets/mixed_realistic_w4_n15_d30.csv")

# Show basic info
df.info()
df.describe()

# Class distribution
sns.countplot(data=df, x="rare_event")
plt.title("Class Balance")
plt.show()

# PCA visualization
pca = PCA(n_components=2)
components = pca.fit_transform(df.drop("rare_event", axis=1))
sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=df["rare_event"])
plt.title("PCA of Feature Space")
plt.show()
```
